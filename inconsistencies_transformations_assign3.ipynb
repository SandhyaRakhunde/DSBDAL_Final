## a) To change the scale for better understanding of the variable : z-score(), minmax()
### data normalization transforms data of varying scales to the same scale. This allows every variable to have similar influence on the model, allowing it to be more stable and increase its effectiveness.

### 1.minmax scaling()

```python
df["Landsize"].head()
```

```python
# 1) using MinMax scaling to "Landsize" => as variation between minimum and maximum value is huge -> scale it down to (0 to 1) range
def minmax_scaling(col):
    return (col-col.min())/(col.max()-col.min())  #using formula

df["Landsize"] = minmax_scaling(df["Landsize"])
df["BuildingArea"] = minmax_scaling(df["BuildingArea"])
df.describe()
```

```python
df["Landsize"].head()
```

### 2. Z-score standardization 

```python
df["Propertycount"].head()
```

```python
# 2) Z-score standardization using pandas 
# The z-score method is often referred to as standardization, value ranges between -3 to 3,
# which transforms the data into a distribution of values where the mean is 0 and has a standard deviation of 1. 

def z_score(col):
    return (col-col.mean())/col.std()

df["Propertycount"] = z_score(df["Propertycount"]).round()
df.describe()
```
