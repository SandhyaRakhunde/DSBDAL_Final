### 4. Compare distributions and identify outliers.
- z-score
- IQR
- MinMax

```python
upper_limit = (df["SepalWidthCm"].mean() + 3* df["SepalWidthCm"].std())
lower_limit = (df["SepalWidthCm"].mean() - 3 *df["SepalWidthCm"].std())

outliers = df[ (df["SepalWidthCm"]<lower_limit) | (df["SepalWidthCm"]>upper_limit) ] 
outliers
```

```python
removed_outliers = df[ (df["SepalWidthCm"]>lower_limit) & (df["SepalWidthCm"]<upper_limit)]
removed_outliers.iloc[10:18]
```
